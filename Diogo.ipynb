## 1. Introdução ao Problema de Estudo
A pandemia de COVID-19, iniciada em 2019, trouxe impactos significativos na saúde pública e gerou uma série de desafios para governos e instituições ao redor do mundo. Uma das principais métricas para avaliar a gravidade da pandemia e sua evolução ao longo do tempo é o excesso de mortalidade — uma medida que representa o número de mortes acima do esperado para um período específico, considerando uma linha de base histórica de mortalidade.

O excesso de mortalidade é um indicador importante porque engloba não apenas as mortes diretamente causadas pelo COVID-19, mas também aquelas que ocorreram como consequência indireta da pandemia, incluindo fatores como a sobrecarga dos sistemas de saúde, a interrupção de tratamentos médicos para outras doenças e o impacto das medidas de distanciamento social. Analisar o excesso de mortes fornece, portanto, uma visão abrangente das consequências da pandemia e ajuda a identificar períodos críticos em que as mortes aumentaram ou diminuíram.

#### Objetivo do estudo
O objetivo deste estudo é analisar padrões de excesso de mortalidade durante a pandemia de COVID-19 utilizando um modelo estatístico conhecido como Modelo Oculto de Markov (HMM, do inglês Hidden Markov Model). O HMM é uma técnica que permite modelar dados temporais, identificando estados ocultos subjacentes que influenciam as observações visíveis. Neste contexto, o modelo HMM nos ajudará a categorizar períodos de tempo em diferentes "estados" de mortalidade, como "alta mortalidade" e "baixa mortalidade", e a observar as transições entre esses estados ao longo do tempo.

## 3. História e Revisão de Literatura do Problema e dos Métodos/Algoritmos

### O Problema do Excesso de Mortalidade Durante Pandemias

Historicamente, o conceito de **excesso de mortalidade** é utilizado para medir o impacto de crises de saúde pública, especialmente durante pandemias. A ideia é calcular a diferença entre o número real de mortes e o número esperado com base em médias históricas. Este indicador é amplamente utilizado por pesquisadores e órgãos de saúde pública porque ele inclui não apenas as mortes diretamente relacionadas a uma doença específica, mas também as consequências indiretas. Durante a pandemia de COVID-19, por exemplo, fatores como sobrecarga hospitalar, adiamento de tratamentos, e até os impactos de medidas de isolamento social contribuíram para um aumento nas taxas de mortalidade além das mortes diretamente causadas pelo vírus.

Estudos como os realizados pela **Organização Mundial da Saúde (OMS)** e pelo **Our World in Data** utilizam o excesso de mortalidade para comparar o impacto da pandemia em diferentes regiões, considerando fatores demográficos e econômicos. Isso ajuda a identificar não apenas as áreas mais afetadas, mas também os períodos de maior vulnerabilidade, permitindo uma análise mais detalhada das respostas e intervenções necessárias. Este conceito foi fundamental em estudos de crises anteriores, como a gripe espanhola de 1918 e pandemias mais recentes, como a H1N1 de 2009, onde ele foi utilizado para avaliar a eficácia das medidas de saúde pública e o impacto geral na população.

### Revisão de Métodos para Análise de Dados Temporais na Saúde Pública

A análise de dados temporais, especialmente no contexto de crises de saúde pública, é fundamental para identificar padrões e mudanças ao longo do tempo. Vários métodos têm sido aplicados para lidar com esses dados, cada um com suas particularidades:

1. **Modelos Estatísticos Tradicionais**: Estes incluem métodos de regressão linear e ARIMA (AutoRegressive Integrated Moving Average), que são populares para séries temporais devido à sua capacidade de prever valores futuros com base em padrões históricos. Porém, esses modelos têm limitações ao lidar com dados complexos e não estacionários, como os de pandemias, onde há variações abruptas e imprevisíveis.

2. **Modelos Baseados em Redes Neurais**: Com o avanço do aprendizado profundo, métodos como redes neurais recorrentes (RNNs) e LSTMs (Long Short-Term Memory) têm sido utilizados para prever séries temporais complexas. Embora esses modelos possam capturar padrões complexos, eles exigem grandes volumes de dados e capacidade computacional significativa, tornando-os menos práticos para algumas aplicações em saúde pública.

3. **Modelo Oculto de Markov (HMM)**: O HMM é uma ferramenta amplamente usada para modelar séries temporais quando há um estado oculto subjacente que influencia as observações visíveis. Originalmente introduzido na década de 1960 por Leonard E. Baum e outros pesquisadores, o HMM é baseado em uma cadeia de Markov onde os estados ocultos não são diretamente observáveis, mas podem ser inferidos através de variáveis observáveis. O HMM tem sido amplamente aplicado em áreas como reconhecimento de fala, análise de sequências biológicas e finanças, além de sua crescente aplicação em saúde pública para identificar estados de “alta” e “baixa” incidência de doenças ao longo do tempo.

4. **Modelos Baseados em Estado**: Além do HMM, modelos como o Dynamic Bayesian Network (DBN) e Markov Switching Models (MSM) são usados para modelar mudanças de estados em séries temporais. Contudo, o HMM é especialmente útil para situações em que os dados são ruidosos e existe a necessidade de inferir estados ocultos não diretamente observáveis.

### O Uso do HMM em Estudos de Saúde Pública

O HMM oferece uma abordagem poderosa para estudos de saúde pública devido à sua capacidade de lidar com **estados ocultos** e **transições probabilísticas** entre estados, que são comuns em dados epidemiológicos. Em estudos de pandemias, o HMM permite identificar períodos de alta e baixa mortalidade (ou incidência de casos), oferecendo uma segmentação automática que pode ser interpretada para inferir o efeito de diferentes políticas públicas ao longo do tempo.

Por exemplo, durante a pandemia de COVID-19, o HMM pode modelar períodos em que o excesso de mortalidade foi elevado e inferir se essas mudanças estão correlacionadas com intervenções de saúde pública, como lockdowns, campanhas de vacinação, ou novas variantes do vírus. Estudos anteriores de epidemias, como o surto de SARS em 2003 e a pandemia de H1N1, usaram HMMs para caracterizar padrões temporais e identificar “ondas” de incidência, demonstrando sua eficácia em contextos de saúde pública.

Em resumo, o HMM é um método bem estabelecido e adequado para a análise de séries temporais epidemiológicas devido à sua:

- **Capacidade de identificar estados ocultos** (como diferentes níveis de mortalidade durante uma pandemia),
- **Modelagem probabilística de transições entre estados**, e
- **Adaptabilidade a dados complexos** e frequentemente ruidosos, comuns em saúde pública.

Essas características tornam o HMM uma escolha metodológica valiosa para o estudo do excesso de mortalidade durante a pandemia de COVID-19, permitindo uma análise estruturada e segmentada que pode informar políticas de saúde pública e gestão de crises em eventos futuros.

```python
import pandas as pd
import numpy as np
from hmmlearn import hmm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid_cleaned_full.csv", low_memory=False)
df
```

```python
# Exibir as primeiras linhas e os nomes das colunas para entender a estrutura do arquivo
print(df.head())
print(df.columns)
```

```python
# Após identificar os nomes corretos, substitua-os aqui:
if 'Cumulative excess deaths per 100,000 people (central estimate)' in df.columns and 'Day' in df.columns:
    # Seleciona as colunas de interesse
    df = df[['Day', 'Cumulative excess deaths per 100,000 people (central estimate)']].dropna()
    df['Day'] = pd.to_datetime(df['Day'], errors='coerce')
    df = df.dropna(subset=['Day'])  # Remove linhas onde a data é inválida
    df = df.sort_values(by='Day')
else:
    print("As colunas esperadas não foram encontradas. Verifique os nomes das colunas no arquivo.")
    exit()
```

```python
# Preparação dos dados para o HMM
excess_deaths = df['Cumulative excess deaths per 100,000 people (central estimate)'].values.reshape(-1, 1)

# Configuração do modelo HMM - modelo com 2 estados (pode ser ajustado)
model = hmm.GaussianHMM(n_components=2, covariance_type="diag", n_iter=100, random_state=42)
```

```python
# Treinamento do modelo
model.fit(excess_deaths)

# Previsão dos estados ocultos
hidden_states = model.predict(excess_deaths)

# Adicionar os estados ocultos ao DataFrame original para análise
df['Hidden State'] = hidden_states
```

```python
# Exibir os parâmetros do modelo para interpretação
print("Matriz de Transição entre Estados:\n", model.transmat_)
print("Médias dos estados:\n", model.means_)
print("Variância dos estados:\n", model.covars_)
```

```python
# Plot dos estados ao longo do tempo
plt.figure(figsize=(14, 8))
for i in range(model.n_components):
    state = (hidden_states == i)
    plt.plot(df['Day'][state], df['Cumulative excess deaths per 100,000 people (central estimate)'][state], '.', label=f"Estado {i}")
plt.legend()
plt.xlabel("Day")
plt.ylabel("Cumulative excess deaths per 100,000 people (central estimate)")
plt.title("Análise HMM dos Excesso de Mortes durante COVID-19")
plt.show()
```

A primeira variável, **"Entity"**, identifica a entidade a que os dados se referem, podendo representar países, regiões ou até grupos de estudo específicos. Esta coluna é textual e inclui exemplos como "Portugal", "Espanha" ou "Brasil", sendo essencial para contextualizar as outras variáveis.

A variável **"Day"** regista o dia ou a data específica para a qual os dados foram recolhidos. Este campo pode estar formatado como uma data (por exemplo, "2021-12-31") ou como um número sequencial, caso represente o número de dias desde um ponto de partida, como o início da pandemia.

Outra variável importante é **"Cumulative excess deaths per 100,000 people (central estimate)"**, que apresenta a estimativa central do número acumulado de mortes em excesso por 100.000 habitantes. Esta métrica refere-se apenas às mortes atribuídas diretamente à COVID-19, sem incluir as mortes indirectas ou mortes em excesso.
